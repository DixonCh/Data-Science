```python
#importing required library

import pandas as pd
import matplotlib.pyplot as plt
```

```python
#reading data
#link to data: http://data.opennepal.net/content/district-coverage-water-supply-and-sanitation-2011

read_rain = pd.read_csv("District Coverage of Water Supply and Sanitation  (2011).csv")
```

```python
#Reading first five datas

read_rain.head()
```

```python
#Data Preprocessing

#We are trying to select only the specific column and rows from the data
```

```python
def pieprocess(col_name,target_row):
    df = pd.DataFrame()

    df[col_name] = read_rain[col_name]
    
    
    df['Sub group'] = read_rain['Sub group']
    df['Value'] = read_rain['Value']
    
    df_idx = df.set_index('Sub group')
    df_swap = df_idx.loc[target_row]
    df_swap = df_swap.groupby(col_name,as_index = True)['Value'].mean()
    plt = df_swap.plot.pie(x= col_name,y='Value', figsize=(5, 5))
    return plt
```

```python
df_swap = pieprocess('Zone','Total Population')
```

```python
df_swap = pieprocess('Zone','Water Coverage Supply (%)')
```
