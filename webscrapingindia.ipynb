```python
import urllib.request

#import the beatiful soup functions to parse the data
from bs4 import BeautifulSoup
```

```python
#specify url
wiki = "https://en.wikipedia.org/wiki/List_of_state_and_union_territory_capitals_in_India"
#wiki = "https://en.wikipedia.org/wiki/List_of_capitals_in_the_United_States"
```

```python
#query the website
page = urllib.request.urlopen(wiki)
```

```python
#parse the html and store in Beautiful soup fromat
soup = BeautifulSoup(page)
```

```python
#soup.find_all("a")

all_links = soup.find_all("a")
for link in all_links:
    print(link.get("href"))
```

```python
all_tables = soup.find_all('table')

right_table = soup.find('table',class_ = "wikitable sortable plainrowheaders")
```

```python
print(right_table)
```

```python
#Generate lists

A = []
B = []
C = []
D = []
E = []
F = []
G = []

for row in right_table.findAll("tr"):
    cells = row.findAll('td')
    
    #To store second column data
    states = row.findAll('th')
    
    #Only extract table body not heading
    
    if(len(cells) == 6):
        A.append(cells[0].find(text = True))
        B.append(states[0].find(text = True))

        C.append(cells[1].find(text = True))
        D.append(cells[2].find(text = True))
        E.append(cells[3].find(text = True))
        F.append(cells[5].find(text = True))
        G.append(cells[5].find(text = True))
```

```python
#Import pandas to convert list to data frame

import pandas as pd

df = pd.DataFrame(A, columns = ['Number'])

df['State/UT'] = B
df['Admin_Capital'] = C
df['Legislative_Capital'] = D
df['Judiciary_Capital'] = E
df['Year_Capital'] = F
df['Former_Capital'] = G
```
